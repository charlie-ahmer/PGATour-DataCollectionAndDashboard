# Exploratory Data Analysis on PGA Player Performance Data
### (Contains data from 2007-2022)

In this notebook, we will perform an Exploratory Data Analysis (EDA) on the PGA player performance dataset obtained from pgatour.com. The dataset contains information about various golf players, including their scoring averages, birdie averages, driving distances, and other performance metrics. The goal of this analysis is to gain insights and understand the patterns, relationships, and trends in the data.

We will begin by loading the dataset and examining its structure and contents. Then, we will apply several EDA techniques to explore the data, such as calculating summary statistics, visualizing distributions and relationships, detecting outliers, and checking for missing data. Additionally, we will perform correlation analysis to assess the relationships between different variables.

Furthermore, we will investigate the change in averages over time for stats such as Scoring Average and Driving Distances. This will help us understand how the player's performance changes over the years.

By the end of this analysis, we aim to uncover valuable insights about golf player performance and understand the factors that contribute to success in the sport.

## Data Dictionary:
Here is a brief description of the columns in the dataset:

| Column Name         | Description                                                                 |
|---------------------|-----------------------------------------------------------------------------|
| PLAYER              | The name of the player.                                                     |
| Country             | The player's home country                                                   |
| Year                | The year in which the data is recorded.                                      |
| ScoringAvg          | The average score of the player.                                             |
| BirdieAvg           | The average number of birdies made by the player per round.                            |
| DrivingDistance     | The average distance of the player's drives.                                 |
| TotalDriving        | The player's total driving score.                                         |
| GIR%                | The percentage of greens in regulation (GIR) achieved by the player.         |
| Scrambling          | The player's scrambling rate, indicating the ability to recover from difficult positions.|
| SG_Total            | The player's strokes gained total.                                           |
| SG_ApproachGreen    | The player's strokes gained on approach shots to the green.                  |
| SG_Putting          | The player's strokes gained through putting.                                 |
| SG_OffTee           | The player's strokes gained off the tee.                                     |
| SG_AroundGreen      | The player's strokes gained around the green.                                |
| SG_TeeToGreen       | The player's strokes gained from tee to green.                               |
| Top10Finishes       | The number of top 10 finishes achieved by the player.                        |
| #Wins               | The number of wins achieved by the player.                                   |
| FedExCupPoints      | The player's total FedEx Cup points.                                         |
| WorldRank           | The player's average points, where higher points indicate a higher world rank.|
| Money               | The amount of money earned by the player.                                    |


```python
##### Install Required Packages #####
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Begin by reading in the csv file obtained through web scraping

df = pd.read_csv('pga_full.csv')
```

```python
df # Show the full data frame
```

```python
# Examine the data frame to identify any missing values
# The csv file obtained from scraping does not contain any null values

df.isnull().sum() 
```

```python
# Calculate summary statistics for numeric features
summary_stats = df[['ScoringAvg', 'BirdieAvg', 'DrivingDistance', 'TotalDriving','GIR%','Scrambling','SG_Total',
                   'SG_ApproachGreen','SG_Putting','SG_OffTee','SG_AroundGreen','SG_TeeToGreen','Top10Finishes',
                   '#Wins','FedExCupPoints','WorldRank','Money']].describe()

# show/examine summary statistics
summary_stats
```

```python
# store the counts of players by country
country_counts = df['Country'].value_counts()

# Create a bar chart
plt.bar(country_counts.index, country_counts.values)
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.title('Distribution of Players by Country')
plt.xticks(rotation=90)
plt.show()
```

Due to the PGA being an American Association, it is not suprising to see the vast majority of golfers being from the USA.

```python
# Store a df of numeric features/columns
df_numeric = df.drop(['PLAYER', 'Year'], axis=1)
numeric_cols = df_numeric.select_dtypes(['int64', 'float64']).columns


# Plot a grid of histograms for each numeric feature for easy examination
f, ax = plt.subplots(nrows=(len(numeric_cols) + 1) // 2, ncols=2, figsize=(18, 18))

for i, column in enumerate(numeric_cols):
    row = i // 2
    col = i % 2
    g = sns.histplot(df_numeric[column], ax=ax[row][col], kde=False)
    g.set(title=column)

plt.subplots_adjust(wspace=None, hspace=1)
plt.show()
```

Each feature is fairly normally distributed aside from Top10Finishes, #Wins, FedExCupPoints, WorldRank, and Money. This makes sense do to the nature of these categories. For example, to have many Top10Finishes, a golfer will need to have stellar performance on the course. These features are competitive, and only a few golfers can obtain big numbers for these columns.

```python
# Plot boxplots for each numeric feature

f, ax = plt.subplots(nrows=(len(numeric_cols) + 1) // 2, ncols=2, figsize=(18, 18))

for i, column in enumerate(numeric_cols):
    row = i // 2
    col = i % 2
    g = sns.boxplot(df_numeric[column], ax=ax[row][col], orient='h')
    g.set(title=column)

plt.subplots_adjust(wspace=None, hspace=1)
plt.show()
```

These boxplots provide a similar conclusion from the last cell. The boxplots, however, allow us to examine the distribution differently to identify outliers.

```python
# This plot will show how Scoring Averages have varied over time

plt.figure(figsize=(6, 3))
sns.lineplot(data=df, x='Year', y='ScoringAvg', marker='o', markersize=6)
plt.xlabel('Year')
plt.ylim(65, 75)
plt.ylabel('Scoring Avg')
plt.title('Scoring Averages on PGA Tour (2017-2022)')
plt.show()
```

Scoring averages have remained fairly similar in the years since 2017.

```python
# This plot will show how Driving Distance varied over time

plt.figure(figsize=(6, 3))
sns.lineplot(data=df, x='Year', y='DrivingDistance', marker='o', markersize=6)
plt.xlabel('Year')
plt.ylim(225, 350)
plt.ylabel('Driving Distance')
plt.title('Driving Distance on PGA Tour (2017-2022)')
plt.show()
```

Driving Distance has had a slight increase since 2017. This could be a result of new driver technology, or golfers increasing focus on their driving performance.

```python
# Create scatter plots
# Each takes a look at how a feature affects a golfer's World Rank
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.scatterplot(data=df, x='WorldRank', y='ScoringAvg')
plt.xlabel('World Rank')
plt.ylabel('Scoring Average')

plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x='WorldRank', y='BirdieAvg')
plt.xlabel('World Rank')
plt.ylabel('Birdie Average')

plt.subplot(1, 3, 3)
sns.scatterplot(data=df, x='WorldRank', y='Money')
plt.xlabel('World Rank')
plt.ylabel('Money')

plt.tight_layout()
plt.show()
```

These plots make sense in that the better a golfer performs (less strokes and more birdies), the higher their World Rank points will be. Also, the best golfers in the world (accoring to World Rank) will also earn the most money.

```python
# Identify the top performers based on ScoringAvg, BirdieAvg, DrivingDistance, and Money
bottom_scoringavg = df.nsmallest(5, 'ScoringAvg')
top_birdieavg = df.nlargest(5, 'BirdieAvg')
top_driving = df.nlargest(5, 'DrivingDistance')
top_money = df.nlargest(5, 'Money')

# Create a function to format the player, value, and year information
def format_top_players(top_df, feature_name):
    top_players = top_df[['PLAYER', feature_name, 'Year']].copy()
    top_players['Rank'] = range(1, len(top_players) + 1)
    top_players = top_players.set_index('Rank')
    return top_players

# Format and display the top players for each feature
top_scoringavg_formatted = format_top_players(bottom_scoringavg, 'ScoringAvg')
top_birdieavg_formatted = format_top_players(top_birdieavg, 'BirdieAvg')
top_driving_formatted = format_top_players(top_driving, 'DrivingDistance')
top_money_formatted = format_top_players(top_money, 'Money')

# Print the top 5 for each feature

print("Top 5 Scoring Average:")
print(top_scoringavg_formatted)

print("\nTop 5 Birdie Average:")
print(top_birdieavg_formatted)

print("\nTop 5 Driving Distance:")
print(top_driving_formatted)

print("\nTop 5 Money:")
print(top_money_formatted)

```

These tables simply show some of the best performances in a variety of categories. it is interesting to note that the top 5 Driving Distance happened from 2020-2022. No golfers from 2017-2019 ended up with higher Driving Distances.

```python
# Simple scatter plot to show the relationshiop between Driving Distance and Birdie Averages
plt.scatter(df["DrivingDistance"], df["BirdieAvg"])
plt.xlabel("Driving Distance")
plt.ylabel("Birdie Average")
plt.title("Relationship between Driving Distance and Birdie Average")
plt.show()
```

This is an interesting plot that shows how Driving Distance affects Birdie Averages. These features appear fairly correlated, in that the farther a player drives the ball, the higher chance they will birdie.

```python
# Calculate the correlation matrix
corr_matrix = df[['ScoringAvg', 'BirdieAvg', 'DrivingDistance', 'TotalDriving','GIR%','Scrambling','SG_Total',
                   'SG_ApproachGreen','SG_Putting','SG_OffTee','SG_AroundGreen','SG_TeeToGreen','Top10Finishes',
                   '#Wins','FedExCupPoints','WorldRank','Money']].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', annot_kws={'size': 8})
plt.title('Correlation Matrix')
plt.show()
```

There are some interesting insights from this correlation matrix:
- Scoring Average has a strong negative correlation with Strokes Gained (SG_...) and Top10Finishes
- SG_Total has a strong positive correlation FexExCupPoints, WorldRank, and Money
- Top10Finishes, #Wins, FedExCupPoints, WorldRank, and Money are all strongly correlated with each other
